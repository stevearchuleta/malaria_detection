<font color=darkblue>

&nbsp;

*Differentiate a Healthy Red Blood Cell from a Malaria-Infected Red Blood Cell.*

*Classify Parasitized Cells from Uninfected Normal Cells by Training a Convolutional Neural Network Model.*

 
Malaria[✳](#malaria), which is an Anopheles mosquito-borne infectious <a id=disease style="text-decoration:none">disease</a> caused by a single-celled microorganism of the Plasmodium group, can severly affect humans beings and cause jaundice, seizures, coma, or death. Transmitted through the bite of infected female mosquitos, malaria is primarily found in tropical regions. The parasite enters a human's blood stream and damages red blood cells. The Plasmodium parasite can stay alive inside a human's body for more than a year without the victim expressing any symptoms. According to the World Health Organization, malaria still <a name="deaths" style="text-decoration:none">kills</a> over 600,000 people annually, almost 2/3rd of whom are under the age of five (5)[✳](#mortality). Until recently, the ability to effectively classify malaria-infected human red blood cells has proven itself to be a laborious endeavor of manually <a name="count" style="text-decoration:none">tallying</a> Plasmodium parasites within a stained microscopic blood smear[✳](#smear). Since 2015, a digital <a name="autoscope" style="text-decoration:none">autoscope</a> has utilized computer vision and classification algorithms to perform identification and quantization of malaria infected blood cells[✳](#quantization); this machine learning task has improved both replication and accuracy.

```python
# ====================================
# STORE PATH OF EXTRACTED TRAIN FOLDER
# ====================================
train_directory = '/content/cell_images/train'

# ================================
# DEFINE IMAGE SIZE FOR UNIFORMITY
# ================================
SIZE = 64

# =================================
# INITIALIZE EMPTY TRAIN_IMAGE LIST
# =================================
train_images = []

# ==================================
# INITIALIZE EMPTY TRAIN_LABELS LIST
# 0 = uninfected, 1 = parasitized
# ==================================
train_labels = []
```

```python
# ===========================================
# RESIZE EACH TRAINNING IMAGE
# CONVERT TRAINING IMAGES TO N-D ARRAYS
# APPEND TRAINING IMAGE TO TRAIN_IMAGE LIST
# CREATE LABELS FOR PARASITIZED IMAGES
# CREATE LABELS FOR UNINFECTED IMAGES
# CONVERT LISTS TO N-D ARRAYS
# ===========================================


for folder_name in ['/parasitized/', '/uninfected/']:
  # ===================
  # PATH OF EACH FOLDER
  # ===================
  images_path = os.listdir(train_directory + folder_name)

  for i, image_name in enumerate(images_path):
    try:
      # ===============
      # OPEN EACH IMAGE
      #================
      image = Image.open(train_directory + folder_name + image_name)
      # =================
      # RESIZE EACH IMAGE
      # =================
      image.resize((SIZE, SIZE))
      # ===========================================
      # CONVERT EACH TRAINING IMAGE TO AN N-D-ARRAY
      # APPEND TO TRAIN_IMAGE LIST
      # ===========================================
      train_images.append(np.array(image))
      # ========================
      # CREATE LABELS FOR IMAGES
      # ========================
      if folder_name == '/parasitized/':
        train_labels.append(1)
      else:
        train_labels.append(0)
    except Exception:
      pass

print(type(train_images))
print(type(train_labels))
# ===========================================
# CONVERT LISTS TO ARRAYS
# ===========================================
train_images = np.array(train_images)
train_labels = np.array(train_labels)

print(type(train_images))
print(type(train_labels))
```

Convolutional Neural Networks possess a property called "translational invariance", wherein the model can still recognize an object as an object even if the appearance varies. Within the training data, it is often advisable to augment the images with the following techniques (if/when appropriate) to introduce and ensure variablity (diversity) within the training dataset. 

1. Horizontal_Flip (T/F) 

2. Vertical_Flit (T/F) 

3. Height_Shift (0 to 1)

4. Width_Shift (0 to 1)

5. Rotation (0 to 180)

6. Sheer (0 to 1) 

7. Zoom (0 to 1)

DO NOT USE DATA AUGMENTATION IN TESTING SET

&nbsp;

Herein, Keras' ImageDataGenerator class is used to automatically augment the training images, which could help improve accuracy in the final model. The ImageDataGenerator class dynamically generates augmented images by using Image Augmentation techniques during training. In order to read the images from a large numpy array, the ImageDataGenerator has three (3) methods: flow(), flow_from_directory(), and flow_from_dataframe().

[🔼](#top)

```python
# ===================================
# ALL IMAGES WILL BE RESCALED 1./255.
# TRAIN IMAGES WILL BE AUGMENTED
# ===================================
# train_datagen = ImageDataGenerator(rescale = 1.0 / 255.0), horizontal_flip = True, vertical_flip = False, height_shift_range = 0.1, width_shift_range = 0.1, rotation_range = 25, shear_range = 0.15, zoom_range = 0.05)
train_datagen = ImageDataGenerator(rescale=1.0 / 255.0,
                              horizontal_flip = True,
                              vertical_flip = False, 
                              height_shift_range= 0.1, 
                              width_shift_range=0.1, 
                              rotation_range=25, 
                              shear_range = 0.15,
                              zoom_range=0.05)
test_datagen = ImageDataGenerator(rescale = 1.0 /  255.0)

# =======================================================================
# FLOW TRAINING IMAGES IN BATCHES OF 20 USING THE TRAIN.DATAGEN GENERATOR
# FLOW TESTING IMAGES IN BATCHES OF 20 USING THE TRAIN.DATAGEN GENERATOR
# FLOW MEANS TO READ IMAGES FROM A LARGE NUMPY ARRAY
# RESIZE IMAGES: 64 x 64
# =======================================================================
train_generator = train_datagen.flow_from_directory(train_dir, batch_size=20, class_mode = 'binary', target_size = (64, 64))
test_generator = test_datagen.flow_from_directory(test_dir, batch_size=20, class_mode = 'binary', target_size = (64, 64))
```
